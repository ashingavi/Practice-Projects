```python
# Importing all necessary libraries required.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
# Seprating the quality data and lable

X = wine_data.drop('quality', axis = 1)
X
```

```python
# Replacing data of quality by 0 and 1 (labe encoding)

Y = wine_data['quality'].apply(lambda Y_value : 1 if Y_value >= 7 else 0 )
Y
```

```python
# spliting data into training and testing data (X and Y)

X_train, X_test, Y_train, Y_test = train_test_split (X, Y, test_size=0.2, random_state=3)
```

```python
print (Y.shape, Y_train.shape, Y_test.shape)
```

```python
# Model Training (Random forest)

model = RandomForestClassifier ()
model.fit (X_train, Y_train)
```

```python
# Model evaluation on our test data
# accurary on test date

X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score (X_test_prediction, Y_test)
print ('Accuracy = ', test_data_accuracy)
```

```python
# Building a predictive system (Wine quality is good or bad)

input_data = (7.3,0.65,0.0,1.2,0.065,15.0,21.0,0.9946,3.39,0.47,10.0)

# changing the input data to a numpy arrary

input_data_as_numpy_array = np.asarray(input_data)

# reshaping the data as we are prediting the lable for only one instance

input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict (input_data_reshaped)
print (prediction)

if prediction == 1:
    print ('It is Good Quality Wine')
else:
    print ('It is Bad Quality Wine')
```
